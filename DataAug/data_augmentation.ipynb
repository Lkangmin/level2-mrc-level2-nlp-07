```python
from pororo import Pororo
import pandas as pd
from datasets import *

pg = Pororo(task="pg", lang="ko")
```

```python
original_dataset = load_from_disk("../../data/train_dataset")
```

```python
original_train_dataset = original_dataset['train']
```

```python
original_train_df = pd.DataFrame(original_train_dataset)

original_train_df
```

```python
para_question = []

for question in original_train_df['question'].tolist():
    para_question.append(pg(question))
```

```python
original_question = original_train_df['question'].tolist()
counter = 0
for i in range(len(para_question)):
    if para_question[i] == original_question[i]:
        counter += 1

print(f"count: {counter}")
```

```python
original_train_df['question'] = para_question

real_original_train_df = pd.DataFrame(original_train_dataset)
```

```python
temp = pd.concat([original_train_df, real_original_train_df], join='outer')

temp
```

```python
original_train_df = temp.drop_duplicates(['question'])
original_train_df
```

```python
original_train_df
```

```python
import pyarrow as pa

paraphrased_train_dataset = Dataset(pa.Table.from_pandas(original_train_df))

paraphrased_train_dataset
```

```python
train_with_no_answer_dataset = load_from_disk('train_with_no_answer')

train_with_no_answer_dataset.features
```

```python
paraphrased_train_dataset.save_to_disk("paraphrased_train")

new_paraphrased_train_dataset = load_from_disk('paraphrased_train')
```

```python
new_paraphrased_train_dataset.features
```

```python
concat_raw = concatenate_datasets(
    [
        new_paraphrased_train_dataset,
        train_with_no_answer_dataset
    ]
)

```
